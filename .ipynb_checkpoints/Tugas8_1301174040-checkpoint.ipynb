```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Load in the dataframe
df = pd.read_csv("https://raw.githubusercontent.com/ahmadrefi/Tugas-8/master/trump_tweets.csv", index_col=0)
```

```python
df.head()
df.drop(columns=['hashtags'])
```

```python
stop_words = ["https", "CO"] + list(STOPWORDS)
```

```python
tweets = " ".join(line for line in df['text'])
```

```python
colors = ["#002868", "#BF0A30"]
colormap = LinearSegmentedColormap.from_list("mycmap", colors)
wc = WordCloud(width = 600, height = 400, random_state=1, background_color='white', colormap=colormap , collocations=True, 
               stopwords = stop_words).generate(tweets)

plt.figure(figsize=[13,13])
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
